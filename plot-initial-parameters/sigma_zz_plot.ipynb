```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for z
z = np.linspace(-30000, 200, 1000)

def calculate_sig_zz(Pf, taper_depth):
    if taper_depth is False: 
        taper_depth = -1000000
    min_sig_zz = 2670.0 * Pf * 9.8 * taper_depth
    sig_zz = np.maximum(min_sig_zz, 2670.0 * Pf * 9.8 * np.minimum(-200.0, z))
    return sig_zz

# Create the plot
plt.figure()

# Plotting for different values of Pf and taper_depth
Pf = 0.25
taper_depth = False
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.25, No floor')

Pf = 0.34
taper_depth = False
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.34, No floor')

Pf = 0.5
taper_depth = False
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.5, No floor')

Pf = 0.34
taper_depth = -6000
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.34, 6km taper')

Pf = 0.34
taper_depth = -8000
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.34, 8km taper')

Pf = 0.34
taper_depth = -10000
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.34, 10km taper')

Pf = 0.5
taper_depth = -10000
sig_zz = calculate_sig_zz(Pf, taper_depth)
plt.plot(sig_zz, z, label='Pf=0.5, 10km taper')

# Reverse the axes
# plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

plt.ylabel('z')
plt.xlabel('sig_zz')
plt.title('sig_zz as a function of z')
plt.legend()
plt.grid(True)
plt.show()
```
