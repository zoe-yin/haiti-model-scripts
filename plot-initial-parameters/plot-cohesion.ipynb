```python
import numpy as np
import matplotlib.pyplot as plt

def cohesion(z):
    cohesion = -0.5e6
    zIncreasingCohesion = 6000.0
    if z >= -zIncreasingCohesion:
        cohesion -= 1.0e6 * (z + zIncreasingCohesion) / zIncreasingCohesion
    return cohesion

# Generate z values from 200 to -30000
z_values = np.linspace(200, -30000, 500)
cohesion_values = [cohesion(z) for z in z_values]

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(z_values, cohesion_values, , label='Cohesion vs. Z')
plt.xlabel('Z')
plt.ylabel('Cohesion')
plt.title('Cohesion as a Function of Z')
plt.gca().invert_xaxis()
plt.grid(True)
plt.legend()
plt.show()
```
