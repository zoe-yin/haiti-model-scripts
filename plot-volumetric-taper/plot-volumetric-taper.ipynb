```python
import numpy as np
import matplotlib.pyplot as plt

# Define the updated tapering function
def create_custom_tapering_array_v2(z, start, end):
    tapering = np.zeros_like(z)
    tapering = np.where(z < end, 1, tapering)
    tapering = np.where((z >= end) & (z <= start), (start - z) / (start - end), tapering)
    tapering = np.where(z > start, 0, tapering)
    return tapering

# Example usage
z = np.linspace(-1500, 1500, 1000)
start = 200  # Set the start value where tapering ends
end = -1000  # Set the end value where tapering begins
tapering_array_v2 = create_custom_tapering_array_v2(z, start, end)

# Plot the results to verify
plt.plot(z, tapering_array_v2, label='Tapering Function')
plt.xlabel('z')
plt.ylabel('Tapering')
plt.title(f'Tapering Function (Decrease from {end} to {start})')
plt.legend()
plt.grid(True)
plt.show()
```
